```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart.csv')
```

```python
data_dup = data.duplicated().any()
print(data_dup)
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.shape
```

```python
data.describe()
```

```python
plt.figure(figsize = (17,6))
sns.heatmap(data.corr(), annot = True)
```

```python
data['target'].value_counts()
```

```python
sns.countplot(x= 'target', data= data, palette = "Set2")
plt.show()
```

```python
data['sex'].value_counts()
```

```python
sns.countplot(x= 'sex', data=data, palette = "Set1")
plt.xticks([0,1],["Females","Males"])
sns
plt.show()
```

```python
sns.countplot(x='sex', hue = 'target', data=data)
plt.xticks([0,1],["Females", "Males"])
plt.legend(labels= ['No Disease', 'Disease'])
plt.show()
```

```python
sns.distplot(data['age'], bins = 20)
plt.show()
```

```python
sns.countplot(x='cp',data=data, palette = "Set3")
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-anginal plan","asymptomatic"])
plt.xticks(rotation = 75)
plt.show()
```

```python
sns.countplot(x="cp",hue="target",data=data,palette ="Set1")
plt.xticks([0,1,2,3],["typical angina","atypical angina","non-anginal plan","asymptomatic"])
plt.legend(labels = ['No Disease', 'Disease'])
plt.show()
```

```python
sns.countplot(x="fbs",hue="target",data=data,palette ="Set2")
plt.legend(labels = ['No Disease', 'Disease'])
plt.show()
```

```python
data['trestbps'].hist()
plt.show()
```

```python
g=sns.FacetGrid(data, hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend(labels=['Males','Females'])
plt.show()
```

```python
data['chol'].hist()
plt.show()
```

```python
cate_val = []
cont_val = []
for column in data.columns:
    if data[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```
